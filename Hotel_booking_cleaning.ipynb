## This is about data cleaning where a row was in text and has to be filled with text with new column


```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("hotel-booking-data.txt",delimiter="\t")
```

```python
df
```

```python
df["Company"].value_counts()
```

```python
df.head()
```

```python
df["Text"]=df["Date"].isnull()
```

```python
df.drop(["Text"],axis=1,inplace=True)
```

```python
df
```

```python
mask=df["Room number"].isnull()
```

```python
mask
```

```python
df["Text"]=np.where(mask,df["Date"],np.nan)
```

```python
df
```

```python
df["Text"].fillna(method="bfill",inplace=True)
```

```python
df.dropna(inplace=True)
```
